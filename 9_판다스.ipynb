### 3. 판다스 데이터프레임 활용(p25)

#### 3.1 데이터프레임의 데이터 집계

```python
# (1) 데이터프레임의 전체 기본 통계 출력
df.describe()
```

```python
# (2) 칼럼값별 데이터 개수 보기
df['반'].value_counts()
```

```python
# (3) 개별 칼럼의 통계값 구하기
print(df['국어'].sum())
print(df['국어'].mean())
```

```python
# (4) 그룹별 통계값 구하기
print(df.groupby('반').sum())
print(df.groupby('반').mean())
```
